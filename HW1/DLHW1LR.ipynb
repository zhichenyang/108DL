```python
import scipy.io as sio ##匯入scipy庫讀取資料是字典格式
import matplotlib.pyplot as plt
import numpy as np

matfn=r'C:\Users\zhich\OneDrive\桌面\DL\HW1\data.mat'
data = sio.loadmat(matfn)
print('x:',data['x'].shape,'y:',data['y'].shape)
print('x:',data['x'].ndim,'y:',data['y'].ndim)
x=data['x']
y=data['y']
plt.figure(figsize=(20,7))
plt.plot(x,y,'o')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)   # 線性迴歸建模
print('coefficient:',model.coef_,'\n','intercept:',model.intercept_)

plt.figure(figsize=(20,7))
plt.plot(x,y,'o',alpha=0.3)
plt.plot(x,model.predict(x),linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

##fitting polynomial regression to the dataset
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)

poly_reg.fit(x_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
print('coefficient:',lin_reg_2.coef_,'\n','intercept:',lin_reg_2.intercept_)

plt.figure(figsize=(20,7))
plt.plot(x, y, 'o',alpha=0.3)
plt.plot(x, lin_reg_2.predict(x_poly),linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
